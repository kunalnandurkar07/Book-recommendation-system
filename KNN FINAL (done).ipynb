```python
number = int(input("Enter number of books to recommend: "))
```

```python
def popularity_based(dataframe, n):
    if n >= 1 and n <= len(dataframe):
        data = pd.DataFrame(dataframe.groupby('title')['rating'].count()).sort_values('rating', ascending=False).head(n)
        result = pd.merge(data, df, on='title')
        del result[result.columns[2]]
        result.drop(['url', 'ISBN','rating_y'], axis = 1, inplace = True)
        result.rename(columns = {"rating_x" : "rating"},inplace=True)
        return result.drop_duplicates()
    return "Invalid number of books entered!"
```

```python
print("Top", number, "Popular books are: ")
popularity_based(recommend, number)
```

```python
def search_unique_places(dataframe, place):
 #   place = place.lower()

    if place in list(dataframe['genre'].unique()):
        return dataframe[dataframe['genre'] == place]
    else:
        return "Invalid Entry"
```

```python
#place = input("Enter the Type: ")
#data = search_unique_places(recommend, place)

#if isinstance(data, pd.DataFrame):
 #   data = popularity_based(data, number)

#data.drop_duplicates()
place = input("Enter the Type: ")
data = search_unique_places(recommend, place)

if isinstance(data, pd.DataFrame):
    data = popularity_based(data, number)
    data = data.drop_duplicates()
else:
    print("The data is not a DataFrame. Unable to drop duplicates.")

# Now 'data' contains a DataFrame without duplicates if it was a DataFrame
```
